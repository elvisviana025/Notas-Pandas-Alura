## 1. Imóveis Residenciais

Objetivo: criar um dataframe com imóveis residenciais apenas

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# retornar lista com somente a coluna 'tipo'
list(dados['Tipo'].drop_duplicates())
```

```python
# criando a mão lista com imóveis residenciais
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
# método isin = retorna valor boleano caso elemento esteja na lista
# retorna Series com boleanos informando se o imóvel é residencial
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
type(selecao)
```

```python
# método de seleção do dataframe = ao receber uma series de valores booleanos, retornará apenas os com índice True
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
# verificando quais são os tipos de imóveis existentes
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
# reformando o índice
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```

## 2. Exportando a Base de Dados

Objetivo: criar um dataframe e exportá-lo para um novo arquivo

```python
# criando um arquivo com base no dataframe obtido
# não pode-se exportar o índice, senão virão dois quando importar, por isso esse parâmetro abaixo
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
