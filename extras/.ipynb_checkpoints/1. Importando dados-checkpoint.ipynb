```python
import pandas as pd
```

# Importando dados

## 1. JSON

```python
# exibe o arquivo JSON
json = open('dados/aluguel.json')
print(json.read())
```

```python
# transforma o arquivo JSON em um "dataframe" e o exibe
df_json = pd.read_json('dados/aluguel.json')
df_json
```

## 2. TXT

```python
# exibe o arquivo TXT
txt = open('dados/aluguel.txt')
print(txt.read())
```

```python
# transforma o arquivo TXT em um "dataframe" e o exibe
df_txt = pd.read_table('dados/aluguel.txt')
df_txt
```

## 3. XLSX

```python
# transforma o arquivo XLSX em um "dataframe" e o exibe
df_xlsx = pd.read_excel('dados/aluguel.xlsx')
df_xlsx
```

## 4. HTML

```python
# transforma o arquivo HTML em um "dataframe" e o exibe
df_html = pd.read_html('dados/dados_html_1.html')
df_html[0]
```
